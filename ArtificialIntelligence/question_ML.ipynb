Le prof de INF1015 a la belle habitude de corriger les intras des élèves de deuxième année en avance. Cependant, exceptionnellement, cette année, il décide de se prendre à la dernière minute. Avec le stress du département de génie logiciel/informatique et les étudiants énervés sur Discord, le prof se perd dans ses fichiers et perd les intras de 10 étudiants!

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_set_etudiants['confidence'] = le.fit_transform(data_set_etudiants['confidence'])
etudiants_a_predire['confidence'] = le.transform(etudiants_a_predire['confidence'])
```

```python
data_set_etudiants.corr()
```

### Dans un premier temps nous allons retenir les données ayant une corrélation elevée avec la donnée à expliquer (passed) et faire un modele linéaire.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
columns = ['hours_study', 'missed_classes', 'gpa', 'does_homework']
X = data_set_etudiants[columns]
Y = data_set_etudiants['passed']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.9, random_state=42)
print(f"Train set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train, Y_train)
```

```python
print(f"Logistic Regression Accuracy: {logistic_model.score(X_test, Y_test)}")
```

```python
Y_passed_pred = logistic_model.predict(etudiants_a_predire[columns])
etudiants_a_predire['passed'] = Y_passed_pred
```
